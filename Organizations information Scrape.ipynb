```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
all_org = []
for i in range(218):
    i = str(i)
    print(i)
    link = "https://www.jas-anz.org/our-directory/certified-organisations?combine=&country%5B0%5D=Australia&location=&standard_selective%5B0%5D=AS%252FNZS%2B4801%253A2001&scope=&page="+i
    try:
        results = requests.get(link)
        soup = bs(results.content, 'html.parser')
        main = 'https://www.jas-anz.org'
        for link in soup.find_all("td", {"class": "views-field views-field-name"}):
            li =link.a['href']
            link = main + li
            print(link)
            all_org.append(link)
    except:
        time.sleep(10)
        results = requests.get(link)
        soup = bs(results.content, 'html.parser')
        main = 'https://www.jas-anz.org'
        for link in soup.find_all("td", {"class": "views-field views-field-name"}):
            li =link.a['href']
            link = main + li
            print(link)
            all_org.append(link)
```
